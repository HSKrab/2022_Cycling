```python
df_color = df[['Date', 'Distance (km)']].copy(deep = True)
df_color['Date'] = df_color['Date'].str[:-6].replace('/', '', regex = True)

conditions = [
    (df_color['Distance (km)'] < 100),
    (df_color['Distance (km)'] >= 100) & (df_color['Distance (km)'] < 150),
    (df_color['Distance (km)'] >= 150)
]

colors = ['blue', 'green', 'orange']

df_color['color'] = np.select(conditions, colors)
df_color
```

```python
import os
import folium
from folium import LayerControl
from gpxplotter import read_gpx_file, create_folium_map, add_segment_to_map
import vincent
import json

the_map = create_folium_map()
for gpx_file in os.listdir('/content/drive/MyDrive/日本一周/gpx files'):
    # Define some properties for drawing the line:
    line_options = {'color': df_color[df_color['Date'] == gpx_file[:8]]['color'].values[0], 
                    'weight': 20, 
                    'opacity': 0.5}

    for track in read_gpx_file('gpx files/' + gpx_file):
        for i, segment in enumerate(track['segments']):
            add_segment_to_map(the_map, segment, line_options = line_options)

    # Create a chart using vincent
    idx = np.argmax(segment['elevation'])

    data = {
        'x': segment['Distance / km'],
        'y': segment['elevation'],
    }

    WIDTH = 400
    HEIGHT = 200

    line = vincent.Line(data, iter_idx="x", width = WIDTH, height = HEIGHT)
    line.axis_titles(x = 'Distance (km)', y = 'Elevation (m)')
    line.x_axis_properties(title_offset = 2)
    line.y_axis_properties(title_offset = -10)
    line_json = line.to_json()
    line_dict = json.loads(line_json)

    popup = folium.Popup(max_width = WIDTH+50, show = False)
    chart = folium.Vega(line_dict, width = WIDTH+50, height = HEIGHT+50)
    chart.add_to(popup)

    marker = folium.Marker(
        location=segment['latlon'][idx],
        popup=popup,
        icon=folium.Icon(icon = 'star'),
    )
    marker.add_to(the_map)

the_map.save('Route.html')

# To display the map in a Jupyter notebook:
#the_map
```
